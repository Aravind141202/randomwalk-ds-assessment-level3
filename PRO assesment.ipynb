```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
b_df=pd.read_csv('books.csv')
bt_df=pd.read_csv('book_tags.csv')
r_df=pd.read_csv('ratings.csv')
```

```python
1.#How many books do not have an original title?
count = b_df['original_title'].isnull().sum()
print(count)
```

```python
#2.How many unique books are present in the dataset ? Evaluate based on the 'book_id' after removing records containing null values for original_title column in [books.csv] and corresponding records in [book_tags.csv] and [ratings.csv]
books_cleaned = b_df.dropna(subset=['original_title'])
book_tags_cleaned = bt_df[bt_df['goodreads_book_id'].isin(b_df['book_id'])]
r_df = r_df[r_df['book_id'].isin(b_df['book_id'])]
unique_books_count = books_cleaned['book_id'].nunique()
print(f"The number of unique books after cleaning is: {unique_books_count}")

```

```python
#3.How many unique users are present in the dataset [ratings.csv] ?
unique_users=r_df['user_id'].nunique()
print(unique_users)
```

```python
#4.How many unique tags are there in the dataset
unique_tag=bt_df['tag_id'].nunique()
print(unique_tag)
```

```python
#5.Which tag_id is the most frequently used ie. mapped with the highest number of books [book_tags.csv] ? (In case of more than one tag, mention the tag id with the least numerical value)â€™
tag_count = bt_df['tag_id'].value_counts()
max_count = tag_count.max()
tag_with_max_count = tag_count[tag_count == max_count].index.min()
print( tag_with_max_count)
```

```python

tag_counts = bt_df['tag_id'].value_counts().nlargest(20)
plt.figure(figsize=(10, 6))
tag_counts.sort_values(ascending=True).plot(kind='barh', color='skyblue')
plt.title('Top 20 Tags by Number of Occurrences')
plt.xlabel('Number of Occurrences')
plt.ylabel('Tag ID')
plt.tight_layout()
plt.show()
```
